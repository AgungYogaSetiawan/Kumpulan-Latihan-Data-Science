### import library

```python
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import pandas as pd
```

### play with data and split the data to training and testing

```python
bc = load_breast_cancer()
bc
```

```python
df = pd.DataFrame(bc.data, columns=bc.feature_names)
df
```

```python
tst = pd.DataFrame(bc.target, columns=['Target'])
tst
```

```python
df.corr()
```

```python
feature = pd.DataFrame(df[['mean radius','mean texture','mean perimeter','mean area','mean smoothness','mean compactness']], columns=['mean radius','mean texture','mean perimeter','mean area','mean smoothness','mean compactness'])
feature
```

```python
# split data
X = feature[['mean radius','mean texture','mean perimeter','mean area','mean smoothness','mean compactness']]
y = bc.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# feature scaling
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
# print(f'X_train: \n{X_train}')
# print(f'X_test: \n{X_test}')
```

### model building

```python
# RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)
print(f'Accuarcy Score: \n{accuracy_score(y_test, y_pred)}')
print(f'Cross Val Score Mean: \n{cross_val_score(rfc, X, y, cv=5).mean()}')
```

```python
# DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
print(f'Accuarcy Score: \n{accuracy_score(y_test, y_pred)}')
print(f'Cross Val Score Mean: \n{cross_val_score(dtc, X, y, cv=5).mean()}')
```

```python
# KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
print(f'Accuarcy Score: \n{accuracy_score(y_test, y_pred)}')
print(f'Cross Val Score Mean: \n{cross_val_score(knn, X, y, cv=5).mean()}')
```

```python
# LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print(f'Accuarcy Score: \n{accuracy_score(y_test, y_pred)}')
print(f'Cross Val Score Mean: \n{cross_val_score(lr, X, y, cv=5).mean()}')
```

```python
# SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
print(f'Accuarcy Score: \n{accuracy_score(y_test, y_pred)}')
print(f'Cross Val Score Mean: \n{cross_val_score(svc, X, y, cv=5).mean()}')
```
