```python
def fizzbuzz():
    for i in range(101):
        # habis dibagi dengan 3 dan 5
        if i % 3 == 0 and i % 5 == 0:
            print('Fizz Buzz')
        elif i % 3 == 0:
            print('Fizz')
        elif i % 5 == 0:
            print('Buzz')
        else:
            print(i)

fizzbuzz()
```
